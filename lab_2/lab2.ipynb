```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = [np.random.normal(size=10000, scale=1, loc=0),
        np.random.normal(size=10000, scale=1, loc=2),
        np.random.normal(size=10000, scale=1, loc=3),
        np.random.normal(size=10000, scale=2, loc=0)]

fig, axs = plt.subplots(1, 4)
fig.set_size_inches(18, 5)
lim = 8
axs[0].set_xlim(-lim, lim)
axs[0].hist(data[0])
axs[1].set_xlim(-lim, lim)
axs[1].hist(data[1])
axs[2].set_xlim(-lim, lim)
axs[2].hist(data[2])
axs[3].set_xlim(-lim, lim)
axs[3].hist(data[3])
```

```python
fig, axs = plt.subplots(1, 4)
fig.set_size_inches(18, 5)
lim = 8
axs[0].set_xlim(-lim, lim)
axs[0].hist(data[0], bins=4)
axs[1].set_xlim(-lim, lim)
axs[1].hist(data[1], bins=4)
axs[2].set_xlim(-lim, lim)
axs[2].hist(data[2], bins=4)
axs[3].set_xlim(-lim, lim)
axs[3].hist(data[3], bins=4)
```

```python
fig, axs = plt.subplots(1, 4)
fig.set_size_inches(18, 5)
lim = 8
axs[0].set_xlim(-lim, lim)
axs[0].hist(data[0], bins=14)
axs[1].set_xlim(-lim, lim)
axs[1].hist(data[1], bins=14)
axs[2].set_xlim(-lim, lim)
axs[2].hist(data[2], bins=14)
axs[3].set_xlim(-lim, lim)
axs[3].hist(data[3], bins=14)
```

```python
fig, axs = plt.subplots(1, 4)
fig.set_size_inches(18, 5)
lim = 8
axs[0].set_xlim(-lim, lim)
axs[0].hist(data[0], cumulative="True")
axs[1].set_xlim(-lim, lim)
axs[1].hist(data[1], cumulative="True")
axs[2].set_xlim(-lim, lim)
axs[2].hist(data[2], cumulative="True")
axs[3].set_xlim(-lim, lim)
axs[3].hist(data[3], cumulative="True")
```

```python
from scipy.special import expit

expit_data = [expit(elem) for elem in data]

fig, axs = plt.subplots(1, 4)
fig.set_size_inches(18, 5)
lim = 2
axs[0].set_xlim(-lim, lim)
axs[0].hist(expit_data[0])
axs[1].set_xlim(-lim, lim)
axs[1].hist(expit_data[1])
axs[2].set_xlim(-lim, lim)
axs[2].hist(expit_data[2])
axs[3].set_xlim(-lim, lim)
axs[3].hist(expit_data[3])
```

```python
fig, axs = plt.subplots(1, 4)
fig.set_size_inches(18, 5)
lim = 1
axs[0].set_xlim(-lim, lim)
axs[0].hist(expit_data[0], cumulative="True")
axs[1].set_xlim(-lim, lim)
axs[1].hist(expit_data[1], cumulative="True")
axs[2].set_xlim(-lim, lim)
axs[2].hist(expit_data[2], cumulative="True")
axs[3].set_xlim(-lim, lim)
axs[3].hist(expit_data[3], cumulative="True")
```
